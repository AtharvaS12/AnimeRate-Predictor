# Train Test Split

```python
df.shape
```

```python
Y = df['Score']
X=df.drop('Score',axis=1)
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=7,
                                                 stratify=Y)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

# Impute outliers from training data

```python
num_col
```

```python
def impute_outliers_z(df_in,col):
    df_in = df_in.copy()
    df_in.loc[df_in[col]<-3,col] = -3
    df_in.loc[df_in[col]>3,col] = 3
    return df_in
```

```python
X_train_z = X_train.copy()
for col in num_col:
    X_train_z = impute_outliers_z(X_train_z,col)
X_train_z.loc[:,num_col].min()
```

```python
X_train_z.loc[:,num_col].max()
```

```python
def impute_outliers_i(df_in,col):
    df_in = df_in.copy()
    q1,q3= df_in[col].quantile([0.25,0.75])
    iqr = q3-q1
    min_val = q1 - 1.5 * iqr
    max_val = q3+1.5 * iqr
    df_in.loc[df_in[col] < min_val,col] = min_val
    df_in.loc[df_in[col] > max_val,col] = max_val
    return df_in
```

```python
X_train_i = X_train.copy()
for col in num_col:
    X_train_i = impute_outliers_i(X_train_i,col)
X_train_i.loc[:,num_col].min()
```

```python
X_train_i.loc[:,num_col].max()
```

# Apply Feature selection using PCA and DT

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99, random_state=7)
pca.fit(X_train_z)
X_train_z_pca = pca.transform(X_train_z)
X_test_z_pca = pca.transform(X_test)
X_train_z_pca.shape, X_test_z_pca.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99, random_state=7)
pca.fit(X_train_i)
X_train_i_pca = pca.transform(X_train_i)
X_test_i_pca = pca.transform(X_test)
X_train_i_pca.shape, X_test_i_pca.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=7)
dt.fit(X_train_z,Y_train)
sel_col = X_train_z.columns[dt.feature_importances_ > dt.feature_importances_.mean()] 

X_train_z_dt = X_train_z.loc[:,sel_col]
X_test_z_dt = X_test.loc[:,sel_col]
X_train_z_dt.shape, X_test_z_dt.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=7)
dt.fit(X_train_i,Y_train)
sel_col = X_train_i.columns[dt.feature_importances_ > dt.feature_importances_.mean()] 

X_train_i_dt = X_train_i.loc[:,sel_col]
X_test_i_dt = X_test.loc[:,sel_col]
X_train_i_dt.shape, X_test_i_dt.shape
```

# Compare different Models on Accuracy F1 score and logloss

```python
from sklearn.metrics import accuracy_score, f1_score, log_loss
res_list =[]
model_names = ['z_pca_lr', 'i_pca_lr', 'z_dt_lr','i_dt_lr',
               'z_pca_rf', 'i_pca_rf', 'z_dt_rf','i_dt_rf',
              ]
pred_list=[Y_pred_z_pca_lr, Y_pred_i_pca_lr, Y_pred_z_dt_lr, Y_pred_i_dt_lr,
          Y_pred_z_pca_rf, Y_pred_i_pca_rf, Y_pred_z_dt_rf, Y_pred_i_dt_rf,]
for i in range(len(pred_list)):
    res = []
    res.append(model_names[i])
    acc = accuracy_score(Y_test,pred_list[i])
    res.append(acc)
    f1 = f1_score(Y_test,pred_list[i],  average='weighted')
    res.append(f1)
    # ll = log_loss(Y_test, pred_list[i])
    # res.append(ll)
    res_list.append(res)
len(res_list)
```
